```python
from astropy.coordinates import Distance
Distance(parallax=df_csv['parallax'][0] * u.mas)
```

```python
l, b, dist = [], [], []

for index, row in df_csv.iterrows():
    p = row['parallax']*u.mas
    if p<0:
        print('negative p')
        continue
    else:
        d = Distance(parallax=p) #, allow_negative=True
        dist.append(d.to_value()) # in parsec 

        eq = SkyCoord(row['ra'], row['dec'], unit=u.deg)
        gal = eq.galactic
        l.append(gal.l.to_value())
        b.append(gal.b.to_value()) # in degrees
#     print(gal.l, gal.b)
dist
```

```python
gal.l
```

```python
len(dist)
```

```python
np.isnan(dist)
```

```python
l = np.array(l)[~np.isnan(dist)]
b = np.array(b)[~np.isnan(dist)]
dist = np.array(dist)[~np.isnan(dist)]
len(l)
```

```python
xlist, ylist, zlist = [], [], []

for i in range(len(dist)):
    _l, _b = l[i]*np.pi/180, b[i]*np.pi/180
    _d = dist[i]
    x = _d*np.cos(_b)*np.cos(_l)
    y = _d*np.cos(_b)*np.sin(_l)
    z = _d*np.sin(_b)
    xlist.append(x)
    ylist.append(y)
    zlist.append(z)
```

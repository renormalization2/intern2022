```python
import os, glob
import imageio

wd = "/home/dhhyun/intern2022/enzo-dev/run/Hydro/Hydro-3D/CollapseTestNonCosmological"
outdir = "/home/dhhyun/intern2022/CollapseTestNonCosmological"

flist = glob.glob(f"{outdir}/DD????.png")
flist = sorted(flist)
imglist = []
for img, t in zip(flist, times):
    img = imageio.imread(img) # plt and imageio imreads have 4 channels, while cv2 has 3
    
    # annotating the time
    text = f'{t.to_value():.2f} kyr'
    org = (150, 100)
    cv2.putText(img, text, org, 
                fontFace = cv2.FONT_HERSHEY_COMPLEX, fontScale = 1.5, color = (250,225,100), thickness=3)
    # png takes 'uint8', while plt returns np.float32
    imglist.append(img)
    
outfile = os.path.join(outdir, 'CollapseTestNonCosmological_timeline.gif')
imageio.mimsave(outfile, imglist, "GIF", duration=0.1) # duration in sec
```

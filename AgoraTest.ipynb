## Saving as PNG Images

```python
import os
import numpy as np
import yt

wd = "/home/dhhyun/intern2022/enzo-dev/run/Hydro/Hydro-3D/AgoraGalaxy"
outdir = "/home/dhhyun/intern2022/AgoraGalaxy"
if not os.path.exists(outdir):
    os.makedirs(outdir)

rho_ex = []
times = []

for i in range(63):
    # Loading Checkpoints sequentially
    file = "DD" + f"{i}".zfill(4)
    path = os.path.join(wd, file, file)
    
    ds = yt.load(path)
    p = yt.ProjectionPlot(ds, "z", ("gas", "density"))
    p.zoom(15.0)
    # p.show()
    
    # Extracting Information
    ad = ds.all_data()
    rho_ex.append(ad.quantities.extrema("density"))
    times.append(ds.current_time.in_units("Myr"))

    outfile = os.path.join(outdir, f"{file}.png")
    p.save(outfile)
    
rho_ex = np.array(rho_ex)
```

```python
import matplotlib.pyplot as plt

timelist = np.array(times.copy())

fig, ax = plt.subplots(dpi=300)
ax.set(
    xlabel="Time (Myr)",
    ylabel="Density ($g/cm^3$)",
    yscale="log",
#     ylim=(1e-32, 1e-21),
)
ax.plot(timelist, rho_ex[:, 0], "-xk", label="Minimum")
ax.plot(timelist, rho_ex[:, 1], "-xr", label="Maximum")
ax.legend()
```

```python
import os, glob
import imageio
import cv2

flist = glob.glob(f"{outdir}/DD????.png")
flist = sorted(flist)
imglist = []

for img, t in zip(flist, times):
    img = imageio.imread(img) # plt and imageio imreads have 4 channels, while cv2 has 3
#     img = cv2.imread(img) # plt and imageio imreads have 4 channels, while cv2 has 3
    
    # annotating the time
#     text = f'{t/1e6:.0f} Myr' # t.to_value()
    text = f'{t.to_value():.0f} Myr'
    org = (150, 100)
    cv2.putText(img, text, org, 
                fontFace = cv2.FONT_HERSHEY_COMPLEX, fontScale = 1.5, color = (250,225,100), thickness=3)
    # png takes 'uint8', while plt returns np.float32
    imglist.append(img)
    
outfile = os.path.join(outdir, 'AgoraRestart_timeline.gif')
imageio.mimsave(outfile, imglist, "GIF", duration=0.1) # duration in sec
```
